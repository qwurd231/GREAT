```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
df = pd.read_csv('metrics_collection.csv')

# 模型列表和顺序
models = df['model_name'].unique()
mt_levels = [1.0, 1.5, 2.0]
exp_levels = [1.00, 1.25, 1.50, 1.75, 2.00]

# 准备绘图
fig, ax = plt.subplots(figsize=(8, 6))

bar_width = 0.35
index = np.arange(len(models))  # 为每个模型预留位置

# 颜色定义
mt_colors = plt.cm.Blues(np.linspace(0.4, 0.8, len(mt_levels)))
exp_colors = plt.cm.Oranges(np.linspace(0.4, 0.8, len(exp_levels)))

# 遍历每个模型，绘制两根堆叠柱子（MT 和 EXP）
for i, model in enumerate(models):
    model_data = df[df['model_name'] == model]

    # 统计 MT 分布
# 以总样本数为基准
    total_count = len(df)

    mt_counts = model_data['mt'].value_counts().reindex(mt_levels, fill_value=0) / total_count * 100
    exp_counts = model_data['exp'].value_counts().reindex(exp_levels, fill_value=0) / total_count * 100


    # 获取 model_data['mt'] 中的所有行并去掉 model_data['exp'] 中有的行
    mt_only = model_data[~model_data['mt'].isin(model_data['exp'])]
    # print(mt_only)
    print(f"Model: {model}")
    print(f"MT Total: {mt_counts.sum()}, EXP Total: {exp_counts.sum()}")

    # 绘制 MT 堆叠柱（左侧）
    bottom = 0
    for j, mt_val in enumerate(mt_levels):
        ax.bar(i - bar_width/2, mt_counts[mt_val], bar_width, bottom=bottom, color=mt_colors[j], label=f'MT {mt_val}' if i == 0 else "")
        bottom += mt_counts[mt_val]

    # 绘制 EXP 堆叠柱（右侧）
    bottom = 0
    for j, exp_val in enumerate(exp_levels):
        ax.bar(i + bar_width/2, exp_counts[exp_val], bar_width, bottom=bottom, color=exp_colors[j], label=f'EXP {exp_val}' if i == 0 else "")
        bottom += exp_counts[exp_val]

# 坐标轴和图例
ax.set_xticks(index)
ax.set_xticklabels(models, rotation=30, ha='right', fontsize=15)
ax.set_ylabel('Percentage (%)', fontsize=20)

ax.set_title('MT & EXP Value Distribution per Model', fontsize=20)
ax.legend(loc='upper right', fontsize=15)

ax.tick_params(axis='y', labelsize=15)  # 调整 y 轴刻度字体大小为15
plt.tight_layout()
plt.savefig('exp&mt_distribution.pdf')
# plt.savefig("model_mt_exp_stacked_bars.png", dpi=300)
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
labels = [
    'exp', 'time', 'nll', 'saccade', 'N100', 'N150', 'N200', 'N250', 'N300',
    'N350', 'N400', 'word_speed', 'line_speed', 'pixel_speed', 'fft',
    'rrt', 'tft', 'skip_rate', 'fixation_time'
]
pearson_r = [
    0.235671, 0.167546, 0.249768, 0.155334, 0.180633, 0.145049, 0.110248, 0.095826,
    0.092410, 0.090664, 0.082864, -0.027614, -0.023530, 0.246473, -0.022425,
    0.003588, 0.037496, -0.013364, -0.103374
]

# 假设分组（你可以自定义）
groups = {
    "Eye-Tracking": ['saccade', 'fixation_time', 'skip_rate'],
    "EEG": ['N100', 'N150', 'N200', 'N250', 'N300', 'N350', 'N400'],
    "Behavioral": ['exp', 'time', 'nll', 'rrt', 'tft'],
    "Speed": ['word_speed', 'line_speed', 'pixel_speed'],
    "Others": ['fft']
}

# 颜色设置
group_colors = {
    "Eye-Tracking": "#FF6F61",
    "EEG": "#6B5B95",
    "Behavioral": "#88B04B",
    "Speed": "#F7CAC9",
    "Others": "#92A8D1"
}

# 构造绘图数据（按 group 顺序）
bar_labels = []
bar_values = []
bar_colors = []
group_names = []

for group, variables in groups.items():
    for var in variables:
        idx = labels.index(var)
        bar_labels.append(var)
        bar_values.append(pearson_r[idx])
        bar_colors.append(group_colors[group])
        group_names.append(group)

# 添加没有在组中的变量
others = set(labels) - set(bar_labels)
for var in others:
    idx = labels.index(var)
    bar_labels.append(var)
    bar_values.append(pearson_r[idx])
    bar_colors.append("#CCCCCC")
    group_names.append("Uncategorized")

# 开始画图
x = np.arange(len(bar_labels))
width = 0.6

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(x, bar_values, color=bar_colors, width=width)

# 加数值标签
for rect, value in zip(bars, bar_values):
    ax.text(rect.get_x() + rect.get_width() / 2, rect.get_height() + 0.01,
            f"{value:.2f}", ha='center', va='bottom', fontsize=9)

# 横线基准线
ax.axhline(0, color='black', linewidth=0.8)

# X轴设置
ax.set_xticks(x)
ax.set_xticklabels(bar_labels, rotation=45, ha='right', fontsize=9)

# Y轴标题
ax.set_ylabel("Pearson r", fontsize=12)
ax.set_title("Correlation of MT with Other Measures", fontsize=14, weight='bold')

# 分组分割线与标签
group_boundaries = []
last = 0
for group in list(groups.keys()) + ["Uncategorized"]:
    count = group_names.count(group)
    if count > 0:
        mid = last + count / 2 - 0.5
        ax.text(mid, max(bar_values) + 0.1, group, ha='center', va='bottom', fontsize=10, weight='bold')
        if last != 0:
            ax.axvline(x=last - 0.5, linestyle='dotted', color='gray')
        last += count

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载数据
df = pd.read_csv('metrics_collection.csv')  # 替换为实际路径

# Step 1: 按 mt 升序排序，然后按 nll 升序排序
df_sorted = df.sort_values(by=['mt', 'nll'], ascending=[True, True]).reset_index(drop=True)

# 获取排序后的数据
x = range(len(df_sorted))
y_left = df_sorted['nll']
y_right = df_sorted['mt']

# 创建图形和主坐标轴
fig, ax1 = plt.subplots(figsize=(10, 6))

# 左轴 - nll
color_left = 'red'
ax1.set_xlabel("排序索引", fontsize=12)
ax1.set_ylabel("nll", color=color_left, fontsize=12)
ax1.plot(x, y_left, color=color_left, label='nll（左轴）', linewidth=1.5)
ax1.tick_params(axis='y', labelcolor=color_left)

# 右轴 - mt
ax2 = ax1.twinx()
color_right = 'blue'
ax2.set_ylabel("mt", color=color_right, fontsize=12)

# 设置 y 轴范围略宽于实际 mt 范围，避免线条太密集
mt_min = y_right.min()
mt_max = y_right.max()
padding = (mt_max - mt_min) * 0.1
ax2.set_ylim(mt_min - padding, mt_max + padding)

ax2.plot(x, y_right, color=color_right, label='mt（右轴）', linewidth=1.2, alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color_right)

# 合并图例
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
plt.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# 标题与美化
plt.title("先按 mt，再按 nll 排序后的趋势图", fontsize=14)
ax1.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取数据
df = pd.read_csv('metrics_collection.csv')  # 替换为实际路径

# 排序数据
df_sorted = df.sort_values(by=['mt', 'nll'], ascending=[True, True]).reset_index(drop=True)

# 初始化变量
x = []
y = []
mt_labels = []
mt_positions = []
current_index = 0
gap = 5  # 每组之间的间隔

# 绘图设置
fig, ax = plt.subplots(figsize=(14, 6))
palette = sns.color_palette('husl', df_sorted['mt'].nunique())

# 每组绘图
for i, (mt_value, group) in enumerate(df_sorted.groupby('mt')):
    n = len(group)
    x_vals = list(range(current_index, current_index + n))
    y_vals = group['nll'].tolist()

    # 平滑处理：滚动平均（window=10 可调）
    y_smoothed = pd.Series(y_vals).rolling(window=10, center=True, min_periods=1).mean()

    # 绘制平滑曲线和填充区域
    ax.plot(x_vals, y_smoothed, color=palette[i], linewidth=3, label=str(mt_value))  # 更粗的线条
    ax.fill_between(x_vals, y_smoothed, alpha=0.25, color=palette[i])

    # 设置标签信息
    mt_positions.append(current_index + n // 2)
    mt_labels.append(str(mt_value))
    current_index += n + gap

# 设置图形样式
ax.set_xlabel("样本索引（按 mt 分组，组内 nll 升序）", fontsize=12)
ax.set_ylabel("nll", fontsize=12)
ax.set_title("不同 mt 分组下 nll 值分布（平滑显示）", fontsize=14)
ax.set_xticks(mt_positions)
ax.set_xticklabels(mt_labels, rotation=45, ha='right')
ax.legend(title="mt 分组", bbox_to_anchor=(1.02, 1), loc='upper left')
ax.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
