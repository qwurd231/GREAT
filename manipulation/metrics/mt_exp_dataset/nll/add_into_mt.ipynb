```python
from typing import List
def load_text_data(path: str) -> List[str]:
    """
    Load text data from a file.
    Args:
        path (str): Path to the file containing text data.
    Returns:
        List[str]: List of text data.
    """
    with open(path, 'r') as f:
        data = f.readlines()
    return data
```

```python
mt_a_conversation = load_text_data('nll_FACEScore/mt-bench-a.txt')
mt_b_conversation = load_text_data('nll_FACEScore/mt-bench-b.txt')
mt_a_sum_nll = load_text_data('mt-bench-a-nll-1.txt')
mt_b_sum_nll = load_text_data('mt-bench-b-nll-1.txt')
```

```python
# calculate sum
for i in range(len(mt_a_sum_nll)):
    mt_a_sum_nll[i] = float(mt_a_sum_nll[i].strip())
    mt_b_sum_nll[i] = float(mt_b_sum_nll[i].strip())
```

```python
import pandas as pd
import numpy as np

mt_bench = pd.read_csv('../mt-bench-ori/mt_bench.csv')
print(mt_bench.head())
```

```python
mt_bench['nll_a'] = mt_a_sum_nll
mt_bench['nll_b'] = mt_b_sum_nll
print(mt_bench.head())
mt_bench.to_csv('../mt_bench.csv', index=False)
```
